```
import pytesseract
import shutil
import os
import random
import cv2
import matplotlib.pyplot as plt
# from google.colab.patches import cv2_imshow
```

## Open Image:

```
img = cv2.imread('sample1.png')
plt.imshow(img)
```

```
crop_img = img[200: 500, 270:790]
plt.imshow(crop_img)
```

```
extractedInformation = pytesseract.image_to_string(crop_img)
```

```
print(extractedInformation)
```

```
crop_img1 = crop_img[30: 270, 10:480]
plt.imshow(crop_img1)
```

```
resize = cv2.resize(crop_img1, (800, 600), interpolation = cv2.INTER_AREA)
plt.imshow(resize)
```

```
extractedInformation = pytesseract.image_to_string(resize)
```

```
print(extractedInformation)
```

## Gray Scale


```
grayImage = cv2.cvtColor(resize, cv2.COLOR_BGR2GRAY)
plt.imshow(grayImage)
```

```
extractedInformation1 = pytesseract.image_to_string(grayImage)
print(extractedInformation1)
```

```
invertImg = cv2.bitwise_not(grayImage)
plt.imshow(invertImg)
extractedInformation2 = pytesseract.image_to_string(invertImg)
print(extractedInformation2)
```

```
(thresh, blackwhite) = cv2.threshold(invertImg, 80, 255, cv2.THRESH_BINARY)
plt.imshow(blackwhite)
extractedInformation3 = pytesseract.image_to_string(invertImg)
print(extractedInformation3)
```
